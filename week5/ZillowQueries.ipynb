```python
import csv, sqlite3

conn = sqlite3.connect(":memory:") # change to 'sqlite:///your_filename.db'
df.to_sql('Zillow_Data', conn, index=False, if_exists='replace')
```

```python
import sqlite3


# Create a cursor object to execute SQL queries.
cursor = conn.cursor()

region_value = 10001
sql_query = f"SELECT * FROM Zillow_Data WHERE RegionName = {region_value} and MedianListingPricePerSqft_3Bedroom > 124"

# Execute the query and read the results into a DataFrame.
df = pd.read_sql_query(sql_query, conn)


# Print the DataFrame
print(df)



```

```python

sql_query = '''
SELECT RegionName, MAX(MedianListingPricePerSqft_3Bedroom) AS MaxMedianListingPricePerSqft_3Bedroom
FROM Zillow_Data
WHERE MedianListingPricePerSqft_3Bedroom IS NOT NULL
GROUP BY RegionName
Order by RegionName;
'''

# Execute the query and read the results into a DataFrame.
result_df = pd.read_sql_query(sql_query, conn)

# Print the result DataFrame
print(result_df)

# Don't forget to close the connection when you're done.
#conn.close()
```
